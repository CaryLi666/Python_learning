```python
# 7-9
sandwich_orders = ['apple_sandwich', 'pastrami', 'banana_sandiwich', 'pastrami', 'potato_sandiwich', 'pastrami']
finished_sandwiches = []
while 'pastrami' in sandwich_orders:
    sandwich_orders.remove('pastrami')
for sandwich in sandwich_orders:
    print('I made your ' + sandwich + ' sandwih.')
    finished_sandwiches.append(sandwich)
    
print('制作好的三明治：', finished_sandwiches)
```

```python
# 7-10
prompt = 'If you could visit one place in the world, where would you go? '
informations = {}
active = True
while active:
    name = input('\nWhat is your name? ')
    places = input(prompt)
    informations[name] = places
    
    # 看是否还有人调查
    repeat = input('Would you like to let annother person respond?(yes/no)')
    if repeat == 'no':
        active = False

# 显示结果
print('-------------------------------------')
for k, v in informations.items():
    print(k + " would like to go " + v + '.')
    
```
